```python
#import pandas and os

import pandas as pd
import os

#import data from new_full_student_data.csv and create DataFrame student_df

full_student_data=os.path.join("Resources","new_full_student_data.csv")

student_df=pd.read_csv(full_student_data)

#verify that data was properly imported with .head()
student_df.head()
```

```python
#check for any NaN data in DataFrame
student_df.isna()
student_df.isna().sum()
#remove NaN data in DataFrame
student_df=student_df.dropna()
#check DataFrame if NaN data has been removed
student_df.isna().sum()
```

```python
#check for duplicate data in DataFrame
student_df.duplicated()
student_df.duplicated().sum()
#drop any duplicated data in student DataFrame
student_df=student_df.drop_duplicates()
#check if duplicated has been dropped
student_df.duplicated().sum()
```

```python
#check data type
student_df.dtypes
#remove "th from 'grade' using str.replace
student_df['grade']=student_df['grade'].str.replace("th","")
#check if th is removed
student_df.head()
```

```python
#change grade data to an int

student_df['grade']=student_df['grade'].astype("int")

#check if data type has been changed

student_df.dtypes

#preview DataFrame

student_df.head(10)
```

```python
#Generate a summary statistics on the data as a whole
student_df.describe()
```

```python
#display the mean math score using the mean()

student_df['math_score'].mean()
```

```python
#Store the minimum reading score as min_reading_score

min_reading_score=student_df['reading_score'].min(0)

#print min_reading_score

min_reading_score
                                            
```

```python
# Select the rows for grade nine and display their summary statistics using `loc` and `describe`.
grade_nine=student_df.loc[student_df['grade']==9]
#print grade_nine with summary statistics
grade_nine.describe()
```

```python
# Store the row with the minimum overall reading score as `min_reading_row`
# using `loc` and the `min_reading_score` found in Deliverable 3.

min_reading_score=student_df['reading_score'].min(0)

min_reading_row=student_df.loc[student_df['reading_score']==min_reading_score]

min_reading_row 
```

```python
# Use loc with conditionals to select all reading scores from 10th graders at Dixon High School.

dix_high_df=student_df.loc[(student_df['grade'] == 10) & (student_df['school_name'] == "Dixon High School"),['school_name','reading_score']]
dix_high_df
```

```python
#find the mean of reading_scores for all students in grade 11 and 12
## using loc and .mean()
reading_mean= (student_df.loc[student_df["grade"]>10,["reading_score"]]).mean()

reading_mean
```

```python
# Use groupby and mean to find the average reading and math scores for each school type.

student_group=student_df.groupby('school_type').mean()

student_group[['reading_score',"math_score"]]

                            

```

```python
# Use the `groupby`, `count`, and `sort_values` functions to find the
# total number of students at each school and sort from most students to least students.

school_count=student_df.groupby("school_name")

student_count=school_count[['student_id']].count()


student_count.sort_values(by="student_id", ascending=False)
```

```python
# use groupby and mean fuction, find the average budger per grade for each school type

budget=student_df.groupby(["school_type","grade"])

budget=budget[["school_budget"]].mean()

budget
                                    
            
```
